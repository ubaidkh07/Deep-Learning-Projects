```python
df['Date']=pd.to_datetime(df['Date'])
df.set_index(df['Date'],inplace=True)
```

```python
close_data=df['Close'].values
```

```python
close_data
```

```python
#reshape into 2-D array with one column
close_data=close_data.reshape(-1,1)
close_data
```

```python
total_records=len(close_data)
```

```python
#80% of data for training
split_perecentage=0.80
split=int(split_perecentage*total_records)
split
```

**Split the data into Training & Testing data**

```python
close_train=close_data[:split]
close_test=close_data[split:]
```

```python
date_train=df['Date'][:split]
date_test=df['Date'][split:]
```

```python
import tensorflow as tf
from keras.models import Sequential     
from keras.layers import LSTM,Dense     
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
look_back=15                        
train_generator=TimeseriesGenerator(close_train,close_train,length=look_back,batch_size=20)
test_generator=TimeseriesGenerator(close_test,close_test,length=look_back,batch_size=20)
```

```python
#define model
model=Sequential()

model.add(LSTM(10,activation='relu',input_shape=(look_back,1)))
model.add(Dense(1))
```

**Compile**

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit_generator(train_generator,epochs=1000)
```

```python
prediction=model.predict_generator(test_generator)
```

```python
model.save('google_stock_predictor.h5')
```

```python
prediction
```

```python
#reshaping back to 1-D
close_train=close_train.reshape((-1))
close_test=close_test.reshape((-1))
prediction=prediction.reshape((-1))
```
